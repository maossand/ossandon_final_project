# MAIN FUNCTION

```python
# items to fetch: 

# Function to get track information
def get_track_info(track_id):
    track_info = sp.track(track_id)
    artist_info = track_info['artists'][0]
    artist = artist_info['name']
    artist_id = artist_info['id']
    title = track_info['name']
    album = track_info['album']['name']
    release_date = track_info['album']['release_date']
    popularity = track_info['popularity']
    explicit = track_info['explicit']
    duration_ms = track_info['duration_ms']  # Duration in milliseconds
    duration_sec = duration_ms / 1000  # Convert to seconds
    audio_features = sp.audio_features(track_id)
    return {
        'artist': artist,
        'artist_id': artist_id,
        'title': title,
        'track_id': track_id,
        'album': album,
        'release_date': release_date,
        'popularity': popularity,
        'explicit': explicit,
        'duration_sec': duration_sec,
        'audio_features': audio_features[0]
    }

# Function to get playlist information
def get_playlist_info(playlist_link):
    # Extract playlist ID from link
    playlist_id = playlist_link.split('/')[-1]

    # Get playlist details
    playlist_info = sp.playlist(playlist_id)
    playlist_title = playlist_info['name']

    # Get tracks in the playlist
    results = sp.playlist_tracks(playlist_id)
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])

    playlist_data = []
    for track in tracks:
        track_id = track['track']['id']
        if track_id:
            track_info = get_track_info(track_id)
            playlist_data.append(track_info)

    return playlist_data
```
