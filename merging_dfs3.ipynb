# Extract Collaborators 2015

```python
#We need to retrieve collaborators from the Artist and Song column


df_ptfk_2015['Collaborator 1'] = None

def extract_and_clean_song(song_title):
    clean_title = song_title
    collaborators = []
    # Update regex to better handle commas and 'and', also trimming empty results
    matches = re.findall(r'\[(ft\.|feat\.) (.*?)\]', song_title, flags=re.IGNORECASE)
    if matches:
        for match in matches:
            # Regex to split by commas or 'and', accounting for optional spaces
            potential_collabs = re.split(r'\s*(?:and|,)\s*', match[1].strip())
            # Filter out any empty strings resulted from split
            collaborators.extend([collab for collab in potential_collabs if collab])
            clean_title = clean_title.replace(f'[{match[0]} {match[1]}]', '').strip()
    return clean_title, collaborators

# Process the 'Artist' column to extract collaborators
for index, row in df_ptfk_2015.iterrows():
    if '/' in row['Artist']:
        parts = row['Artist'].split(' / ')
        df_ptfk_2015.at[index, 'Artist'] = parts[0]
        if len(parts) > 1:
            df_ptfk_2015.at[index, 'Collaborator 1'] = parts[1]

# Initialize columns for song collaborators
df_ptfk_2015['Collaborator 2'] = None
df_ptfk_2015['Collaborator 3'] = None
df_ptfk_2015['Collaborator 4'] = None
df_ptfk_2015['Collaborator 5'] = None

# Process the 'Song' column to extract and clean collaborators
for index, row in df_ptfk_2015.iterrows():
    new_song_title, collabs = extract_and_clean_song(row['Song'])
    df_ptfk_2015.at[index, 'Song'] = new_song_title
    for i, collab in enumerate(collabs, start=2):  # Start assigning from Collaborator 2
        if i <= 5:  # Ensure we only fill up to Collaborator 4
            df_ptfk_2015.at[index, f'Collaborator {i}'] = collab

#index66 neeed to change collaborator 3 to Francis and the Lights and erase what is in collaborator 4

df_ptfk_2015.head()
```

#### Define function to clean

```python
#Get rid of these characters: : , [, ] " '

df_ptfk['Song'] = df_ptfk['Song'].str.replace('"', '', regex=False)

chars_to_strip = """ “:,[]\”'" """ 


columns_to_clean = ['Artist', 'Song', 'Reviewer' 'Collaborator 1', 'Collaborator 2', 
                    'Collaborator 3', 'Collaborator 4', 'Collaborator 5']

for column in columns_to_clean:
    if column in df_ptfk.columns:
        df_ptfk[column] = df_ptfk[column].str.strip(chars_to_strip)
    
```

```python
#in reviewer replace 'Billie' by 'Braudie Blais-Billie'
# 'Chris O‚ÄôConnell' by 'Chris O-Connell'
# 'Jill Mapes' by 'Jillian Mapes'

# Replace the values
replacement_dict = {
    'Billie': 'Braudie Blais-Billie',
    'Chris O‚ÄôConnell': 'Chris O-Connell',
    'Jill Mapes': 'Jillian Mapes',
    'Wood' : 'Brady Brickner-Wood'    
}
df['Reviewer'] = df['Reviewer'].replace(replacement_dict, regex=True)
```

## Create a column with a key year_rank and is a collaboration or not 



```python
# Creating a new column by joining 'Column1' and 'Column2' with an underscore
df_ptfk['year_rank'] = df_ptfk['Year'].astype(str) + '_' + df_ptfk['Ranking'].astype(str)

# Display the DataFrame to verify the new column
print(df_ptfk)
```

```python
# Creating a new column by joining 'Column1' and 'Column2' with an underscore
collaborator_columns = ['Collaborator 1', 'Collaborator 2', 'Collaborator 3', 'Collaborator 4', 'Collaborator 5']

# Create 'is_collaboration' column based on the presence of data in collaborator columns
df_ptfk['is_collaboration'] = df_ptfk[collaborator_columns].apply(
    lambda row: 1 if any(row[col] not in [None, ''] for col in collaborator_columns) else 0, axis=1)
```
